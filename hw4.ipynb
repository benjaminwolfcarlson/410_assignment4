```python
# Step 1
import pandas as pd
df = pd.read_csv('hw4_exams.csv')
df.head()

# Step 2
df.info()

# Step 3
df[['math score', 'reading score', 'writing score']].describe()

# Step 4
df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Step 5
df[['math score']]

# Step 6
df['math score']

# Step 7
df.lunch

# Step 8
df[(df['gender'] == 'female') & (df['math score'] >= 90)]

# Step 9
df[['math score', 'reading score', 'writing score']].plot(kind='box')

# Step 10
gender_avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
gender_avg_scores.plot(kind='bar')

# Step 11
import seaborn as sns
sns.catplot(x='gender', kind='count', data=df)

# Step 12
import matplotlib.pyplot as plt
sns.catplot(x='gender', kind='count', data=df)
plt.xticks(rotation=45)

# Step 13
sns.scatterplot(x='reading score', y='writing score', hue='test preparation course', data=df)

# Step 14
df.plot.scatter(x='reading score', y='writing score', c=df['test preparation course'].map({'none': 'red', 'completed': 'blue'}))

# Step 15
df.plot.scatter(x='reading score', y='writing score', c=df['test preparation course'].map({'none': 'red', 'completed': 'blue'}), figsize=(10, 6))
```
